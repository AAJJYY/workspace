```python
# 실습 3

URL = 'https://comic.naver.com/webtoon/weekday?order=User'
res = requests.get(URL)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
results = soup.find_all(class_='thumb')
```

```python
webtoon_list = []
for result in results:
    a_tag = result.find('a')
    img_tag = result.find('img')
    
    href = a_tag.attrs['href']
    alt = img_tag.attrs['alt']
    webtoon_list.append({
        '제목': alt,
        '링크': href
    })

print(webtoon_list)
```

```python
webtoon_list = [ {'제목': result.find('img').attrs['alt'], '링크': result.find('a').attrs['href'] } for result in results]
# [<표현식> for i in range(20)]
print(webtoon_list)
```

- 서울 창업 디딤터 공지사항 크롤링

```python
URL = 'http://didimteo.or.kr/'
#      'http://didimteo.or.kr/news/notice/'

response = requests.get(URL)
print(response.status_code)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
board_result = soup.select('#kboard-default-latest')[1]
# == 같다
board_result = soup.select_one('#kboard-default-latest')
```

```python
# resultSet, tag 구분 !
board_result[0].text
```

```python
results = board_result.select('tbody > tr')

final_result = []

for result in results:
    # print(result.text)
    # title = result.select('td')[0]
    # date = result.select('td')[1]
    
    title = result.select_one('.kboard-latest-title').text # 제목
    date = result.select_one('.kboard-latest-date').text # 날짜, 시간
    # print(title, date)
    temp = {
        '제목': title,
        '날짜': date
    }
    
    final_result.append(temp)
```

```python
# 문자열에서 필요없는 N 제거
final_result[0]['제목'].replace('N', '').strip()
```

```python
for final_r in final_result: # final_r { 제목: v1, 날짜: v2 }
    # print(final_r)
    # print()
    final_r['제목'] = final_r['제목'].replace('N', '').strip()
    print(final_r)
    print()
# print(final_result)
```

```python
# 고정 (템플릿처럼 사용)
URL = 'http://didimteo.or.kr/'
response = requests.get(URL)
soup = BeautifulSoup(response.text, 'html.parser')


# 구조 분석 후 큰 틀에서 작은곳으로 순차적 접근
board_result = soup.select('#kboard-default-latest')[0]
results = board_result.select('tbody > tr')

# 우리가 필요한 데이터를 담아줄 곳
final_result = []
for result in results:
    title = result.select_one('.kboard-latest-title').text # 제목
    date = result.select_one('.kboard-latest-date').text # 날짜, 시간
    temp = {
        '제목': title,
        '날짜': date
    }
    
    final_result.append(temp)
```
